There are a lot of null values in the rating column

```python
#can we create new features from display_inches, resolution_height and resolution_width??
#total_pixels
#ppi
#aspect_ratio
df['ppi']=np.sqrt(df['resolution_width']**2+df['resolution_height']**2)/(df['resolution_width']*df['resolution_height'])
```

```python
#is there any correlation between ppi, price or rating?
sns.heatmap(df[['ppi','price','rating']].corr(),annot=True, cmap='magma', fmt=".2f")
```

There is a negative correlation between (ppi and price) and (ppi and rating), indicating that as the ppi increases, both the price and rating tend to decrease.

```python
#Does the core count affect the price or the rating of laptops?
plt.figure(figsize=(12,6))
sns.boxplot(df,x='core_count',y='price')
plt.title("Boxplot of core_count in laptops")
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(df,x='core_count',y='rating')
plt.title("Boxplot of core_count in laptops")
```

```python
sns.heatmap(df[['core_count','price','rating']].corr(),annot=True)
```

We observe a trend suggesting that laptops with a higher number of cores tend to have higher prices and ratings. However, it's essential to note that correlation does not imply causation, and further analysis, especially considering the presence of null values in the rating column, is needed to confirm this relationship conclusively.

```python
#Does cpu_gen have any affect on price or rating
sns.heatmap(df[['cpu_gen','price','rating']].corr(),annot=True)
plt.title("Correlation Heatmap of CPU Generation, Price, and Rating")
```

CPU_gen has little or no effect on price as well as the rating of laptops!

```python
#Are there any specific combinations of GPU and CPU that are associated with higher-priced or higher-rated laptops?
cpu_gpu=df.groupby(['gpu_brand', 'cpu_brand']).agg({'price': 'mean', 'rating': 'mean'})
```

```python
cpu_gpu=cpu_gpu.sort_values(by='price',ascending=False)
```

```python
cpu_gpu['price'].plot(kind='bar')
plt.xlabel('GPU-CPU Combinations')
plt.ylabel('Average Price')
plt.title('Top GPU-CPU Combinations Associated with Higher-Priced Laptops')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Yes! NVIDIA and Intel are gpu-cpu combination which is costliest

```python
cpu_gpu['rating'].plot(kind='bar')
plt.xlabel('GPU-CPU Combinations')
plt.ylabel('Average Rating')
plt.title('Top GPU-CPU Combinations Associated with Higher-Rated Laptops')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#Are there any specific combinations of RAM and ROM that are associated with higher-priced or higher-rated laptops?
ram_rom=df.groupby(['ram_type', 'rom_type']).agg({'price': 'mean', 'rating': 'mean'})
ram_rom=ram_rom.sort_values(by='price',ascending=False)
```

```python
ram_rom['price'].plot(kind='bar')
plt.xlabel('RAM-ROM Combinations')
plt.ylabel('Price')
plt.title('Top RAM-ROM Combinations Associated with Higher-Priced Laptops')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
ram_rom['rating'].plot(kind='bar')
plt.xlabel('RAM-ROM Combinations')
plt.ylabel('Average Rating')
plt.title('Top RAM-ROM Combinations Associated with Higher-Rated Laptops')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
